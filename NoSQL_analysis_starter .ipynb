## Notebook Set Up

```python
from pymongo import MongoClient
import pandas as pd
from pprint import pprint
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
db = mongo['uk_food']
```

```python
# review the collections in our database
print(db.list_collection_names())
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Define the query
query = {"scores.Hygiene": 20}

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(f"Number of establishments with a hygiene score of 20: {count}")

# Find the establishments with a hygiene score of 20
hygiene_score_20 = establishments.find(query)

# Display the first document in the results using pprint
print("First establishment with a hygiene score of 20:")
try:
    pprint(next(hygiene_score_20))
except StopIteration:
    print("No establishments with a hygiene score of 20 found.")
```

```python
# Convert the result to a Pandas DataFrame
df_hygiene_20 = pd.DataFrame(list(hygiene_score_20))
# Display the number of rows in the DataFrame
print(f"Number of rows: {len(df_hygiene_20)}")
# Display the first 10 rows of the DataFrame
print(df_hygiene_20.head(10))
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {"LocalAuthorityName": {"$regex": "London", "$options": "i"}, "RatingValue": {"$gte": "4"}}

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(f"Number of establishments in London with a RatingValue >= 4: {count}")

# Display the first document in the results using pprint
london_high_rating = establishments.find(query)
print("First establishment in London with a high rating:")
try:
    pprint(next(london_high_rating))
except StopIteration:
    print("No establishments in London with a high rating found.")
```

```python
# Convert the query result to a DataFrame
df_london_high_rating = pd.DataFrame(list(london_high_rating))

# Display the number of rows in the DataFrame
print(f"Number of rows: {len(df_london_high_rating)}")

# Alternatively, using shape attribute (shape[0] gives number of rows, shape[1] gives number of columns)
print(f"Number of rows: {df_london_high_rating.shape[0]}")

# Display the first 10 rows of the DataFrame
print(df_london_high_rating.head(10))
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Define the coordinates for "Penang Flavours"
latitude = 51.49014200
longitude = 0.08384000

# Query
query = {
    "RatingValue": "5",
    "geocode": {
        "$geoWithin": {
            "$box": [
                [longitude - 0.01, latitude - 0.01],
                [longitude + 0.01, latitude + 0.01]
            ]
        }
    }
}
sort = [("scores.Hygiene", 1)]
limit = 5
results = establishments.find(query).sort(sort).limit(limit)
df = pd.DataFrame(list(results))
print(df.head(5))

# Aggregation pipeline
pipeline = [
    {"$match": {"scores.Hygiene": 0}},
    {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}},
    {"$sort": {"count": -1}},
    {"$limit": 10}
]
results = establishments.aggregate(pipeline)
df = pd.DataFrame(list(results))
print(df.head(10))
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
pipeline = [
    {"$match": {"scores.Hygiene": 0}},
    {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}},
    {"$sort": {"count": -1}}
]

# Execute the pipeline
results = establishments.aggregate(pipeline)

# Convert the results to a list
results_list = list(results)

# Print the number of documents in the result
print(f"Number of documents: {len(results_list)}")

# Print the first 10 results
print("First 10 results:")
for result in results_list[:10]:
    pprint(result)
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(results_list)

# Display the number of rows in the DataFrame
print(f"Number of rows: {len(df)}")

# Display the first 10 rows of the DataFrame
print(df.head(10))
```
